```python
#https://www.statology.org/convert-datetime-to-date-pandas/
#https://stackoverflow.com/questions/39992411/to-datetime-value-error-at-least-that-year-month-day-must-be-specified-pand
#https://stackoverflow.com/questions/30222533/create-a-day-of-week-column-in-a-pandas-dataframe-using-python
#https://stackoverflow.com/questions/41138190/how-to-extract-the-year-month-and-day-in-pandas


#Cleaning df

df = df[df.isna() == False]
df['start_time'] = pd.to_datetime(df['start_time'])
df['end_time'] = pd.to_datetime(df['end_time'])
df['duration_min'] = (df.duration_sec /60).astype(int)
df['duration_hour'] = (df.duration_min /60)
df['year_birth'] = 2022 - df['member_birth_year']
df['start_time_dweek'] = df['start_time'].dt.day_name()
df['end_time_dweek'] = df['end_time'].dt.day_name()
df['start_time_day'] = df['start_time'].dt.day
df['end_time_day'] = df['end_time'].dt.day
df = df[df['member_birth_year']>=1950]

df.head(20)
```

```python
df['member_birth_year'] = df['member_birth_year'].astype('Int64') #df['member_birth_year'].astype('category')
df['decade'] = ((df['member_birth_year']/10).astype('int64')*10).astype(str)
df.info()
```

## Multivariate Exploration

```python
g = sb.FacetGrid( data = df.sample(300), hue = 'member_gender', height = 4, aspect = 1.5 )
g.map(sb.regplot,'year_birth', 'duration_min', x_jitter = 0.04, fit_reg = False);
plt.xlabel('Age')
plt.ylabel('Minutes');
plt.title('Minutes of use per Age grouped by Gender');
plt.legend(title='Gender');
```

```python
#https://stackoverflow.com/questions/37278654/boxplots-by-group-for-multivariate-two-factorial-designs-using-matplotlib-pand

station_start = df.start_station_id.value_counts().iloc[:10]
df_station = df.loc[df['start_station_id'].isin(station_start.index)]
order = ['1950','1960','1970','1980','1990','2000']
sb.catplot(data=df_station.sample(200), x='decade', y='duration_min', 
              hue='member_gender', 
             kind='box', legend=False, order = order)
plt.xlabel('Decade')
plt.ylabel('Minutes');
plt.title('Most common 10 station start by User Decade x Duration per Gender');
plt.legend(title='Gender');
```

```python
station_start = df.start_station_id.value_counts().iloc[:10]
df_station = df.loc[df['start_station_id'].isin(station_start.index)]
g = sb.FacetGrid( data = df_station, hue = 'member_gender', height = 4, aspect = 1.5 )
g.map(sb.regplot,'year_birth', 'duration_min', x_jitter = 0.04, fit_reg = False,scatter_kws = {'alpha' : 1/2});
plt.xlabel('Age')
plt.ylabel('Minutes');
plt.title('Most common 10 station start by User Age x Duration per Gender');
plt.legend(title='Gender');
```
